```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./community.jpg')
## associates img variable with the file in the current folder
```

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
## imports the face recognizer algorithm
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

## converts the image to gray scale, which helps the algorithm
```

```python
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()

## gray image
```

```python
faces = face_classifier.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=5)

## faces will be a tuple with coordinates of where the faces where found
## scaleFactor: higher values - faster but less accurate
## minNeighbors: lower values - detects more faces but may include things that aren't faces
```

```python
if len(faces) == ():
    print("Rostos não encontrados")
## if there's no coordinates of faces then no face was recognized

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (127, 0, 255), 2)
    ## this will draw rectangles where the faces where found
    ## so we can see if it found in the correct places

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
## converts the image from BGR to RGB scale, so it looks like the
## default image

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
